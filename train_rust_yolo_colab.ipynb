### Step 3: Train the Model
We will use the **YOLOv11 Nano** model (yolo11n.pt) because it is very fast and perfect for real-time inference on your laptop CPU.

```python
from ultralytics import YOLO

# Load a pretrained YOLO11n model
model = YOLO("yolo11n.pt")

# Train the model
results = model.train(
    data="./dataset/data.yaml",  # Path to your dataset yaml
    epochs=50,                  # Adjust based on your dataset size
    imgsz=640,                  # Image size
    device=0                    # Use GPU on Colab
)
```
